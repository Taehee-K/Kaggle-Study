```python
# axis = 0: 데이터 방향으로 합침 (위아래)
# axis = 1: 데이터 양옆으로 합쳐짐 
alldata = pd.concat([train, test], axis = 0) 
alldata
# train test 합치면 전처리 1번만 하면 됨
# test 데이터 부분의 정답 column 'IsBadBuy' test 데이터 부분에도 새롭게 생성, none 값으로 설정
```

```python
# RefId : id column 
# IsBadBuy : 정답 column 지우기
# PurcDate : 날짜 column 학습 불가 -> 지우기
alldata2 = alldata.drop(columns = ['RefId', 'IsBadBuy', 'PurchDate']) 
alldata2
```

```python
## categorical 변수 바꾸어주기(문자 column 숫자로)
alldata2['Auction'].unique() # Auction column에 무슨 카테고리가 있는지 확인
```

```python
# 무엇을 무엇으로 바꿀 때: dictionary {} 형식 활용
alldata2['Auction'] = alldata2['Auction'].replace({'ADESA':0, 'OTHER':1, 'MANHEIM':2})
alldata2 #Auction 잘 바뀐 것 확인할 수 있음
```

```python
alldata['Make'].unique() # 종류가 너무 많음 -> 자동으로 바꾸는 것 있었으면 좋겠다 
```

Encoding(인코딩): 문자를 숫자로 바꾸는 과정     
LabelEncoder 라이브러리 가져와서 처리하기   
    1. 어떤 값에 숫자를 등록하겠다 >> fit       
    2. 실제로 등록된 것 바탕으로 변환  >> transform     
fit_transform: 등록과 함게 변환

```python
# 전처리중 >> preprocessing
from sklearn.preprocessing import LabelEncoder 

# 각 category 안의 값들을 알파벳 순서로 정렬, 순서대로 값 부여
le = LabelEncoder() #Label Encoder 함수 정의하기
alldata2['Make'] = le.fit_transform(alldata2['Make'])
alldata2 # Make column int 형으로 변환됨
```

```python
le.classes_ # Label Encoder의 classes 번호 부여 기준 (알파벳 순서대로 0부터)
# 트리 모델 -> 어떤 column이 카테고리적 변수인지 알아서 처리
# One Hot Encoding 시키면 column 너무 많아짐 (KNN 같은 경우는 해줘야 함)
```

```python
alldata2.dtypes ==object
```

```python
# 문자형 column에 대해 label encoding 해주기 -> for 문으로 한꺼번에 처리
# 
# 전체 column 중 categorical 변수들만 접근
# type int, float 아닌 columns들
cat = alldata2.columns[alldata2.dtypes ==object] # column 중 데이터타입이 object인 것들만
cat
```

```python
for col in cat:
#     print(col)
    alldata2[col] = le.fit_transform(alldata2[col]) # 각 column 돌아가면서 encode
alldata2
```

train, test를 따로따로 fit_transform 적용할 때 
-> 알파벳 순서로 되다 보니 train/test에 있는 종류의 개수가 완벽히 같지 않을 수 있음
-> 처리가 될 때 알파벳 순서로 정렬 후 카테고리 번호를 부여하다 보니 같은 숫자로 매핑되지 않을 수 있다
-> 아예 처리시 alldata로 처리하는 것이 현명

<결측치>        
- None 값, 다른 값과는 다른 카테고리로 학습함
- None 값의 의미에 따라 다르게 다뤄야 함 (배경지식이 있을 때)
- 배경지식이 없을 때 >> 부스팅 모델     
    + 모델이 알아서 결측치를 만났을 때 결측치에 어떤 값이 들어가야지 판단   
    + 내부적으로 가장 점수 계산이 많이 되는 방향으로    
    + but overfitting 될 가능성 있음        
안전한 방법: 결측치 타 값들과 아예 다른 값으로 채우기

```python
alldata2 = alldata2.fillna(-1) # 결측치 -1로 채워짐
alldata2
```

```python
# iloc 사용시: alldata2.iloc[데이터, 열]
train2 = alldata2[:len(train)] # 맨 앞의 데이터부터 트레인셋의 개수까지 가져옴
test2 = alldata2[len(train):]

display(train2.head(), test2.head())
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_jobs = -1) # njobs = -1 모든 cpu 병렬로 활용
rf.fit(train2, train['IsBadBuy'])
# result = rf.predict(test2)

# 99%의 데이터 -> 정답, 1번일 때의 확률의 더 중요
# IsBadBuy > 사기를 당할 확률이 더 중요
# 각 데이터별로 확률값, 정도에 대해 분류
# classification -> 분류 기준, threshold 조정 들 고려하
result = rf.predict_proba(test2)
result
```

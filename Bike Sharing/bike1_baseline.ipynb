```python
train = pd.read_csv("/kaggle/input/bike-sharing-demand/train.csv")
test = pd.read_csv("/kaggle/input/bike-sharing-demand/test.csv")

# train, test 데이터 한번에 보기
display(train, test)
```

```python
# 데이터 전처리 -> train/test column 수 맞추기, 변수형 float/int로 맞추기 등
# !!!!! 최소한의 데이터 전처리만으로 baseline 설정 후 전처리 진행하기

# train -> casual+registered = count 
train2 = train.drop(columns = ['casual', 'registered', 'count', 'datetime'])
test2 = test.drop(columns = 'datetime')
display(train2, test2)
```

```python
## Modeling
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_jobs = 4)# 모델 선언
rf.fit(train2, train['count'])# 모델 학습
result = rf.predict(test2)# 모델 예측
result
```

```python
sub = pd.read_csv("/kaggle/input/bike-sharing-demand/sampleSubmission.csv")
sub.head()
```

```python
sub['count'] = result
sub.to_csv("submission_baseline", index = False)
```
